```python
#词频统计及排序

import os
os.listdir(r'D:/Echo/Advanced Python/corpara')                    #获取文件夹下的所有文件名


for info in os.listdir(r'D:\Echo\Advanced Python\corpara'):
    domain = os.path.abspath(r'D:\Echo\Advanced Python\corpara')                 #获取文件夹的路径
    info = os.path.join(domain,info)                            #将路径与文件名结合起来就是每个文件的完整路径
    info = open(info,'r',encoding="utf-8").read()                                     #读取文件内容
    #print(info)    
    new_info = open('D:/Echo/Advanced Python/corpara/all.txt','a+',encoding="utf-8")             #创建新的文件
    new_info.write(info)                                                        #将文件内容循环写入新文件
    
    
new_info = open(r'D:/Echo/Advanced Python/corpara/all.txt','r',encoding="utf-8")
new_info = new_info.read()   
words = new_info.replace(",","").replace("."," ").replace(":","").replace(";","").replace("?","").replace('"'," ").replace("!","").replace("&"," ")
words = words.lower()
wordlist = words.split()
#print(wordlist)
  
word_sort = {}
for word in wordlist:
    num = wordlist.count(word)
    word_sort[word] = num
    
#print (word_sort)
#print(word_sort['the'])    
word_most = sorted(word_sort.items(),key=lambda e:e[1],reverse = True)

#print(word_most)

print("词频前十：")
print("freq*\tword")
print("-----\t-----")


i=0
for key in word_most:
    print(str(key[1])+"\t"+key[0])
    i += 1
    if(i == 10):
        break;
         
print('\n')        
print("词频倒数前十：")
print("freq*\tword")        
print("-----\t-----")

word_least = sorted(word_sort.items(),key=lambda e:e[1],reverse= False)

n=0
for key in word_least:
    print(str(key[1])+"\t"+key[0])
    n += 1
    if(n == 10):
        break;

#info.close()

    
```
